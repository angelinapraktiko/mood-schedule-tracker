```python
# Sistem Program Perencanaan Jadwal Kuliah dan Mood Harian

import datetime

class KegiatanKuliah:
    def __init__(self, nama_matkul, hari, jam_mulai, jam_selesai):
        self.nama_matkul = nama_matkul
        self.hari = hari
        self.jam_mulai = jam_mulai
        self.jam_selesai = jam_selesai

    def __str__(self):
        return f"{self.nama_matkul} ({self.hari}, {self.jam_mulai}-{self.jam_selesai})"

class MoodHarian:
    def __init__(self):
        self.data_mood = {}  # key: tanggal, value: mood

    def catat_mood(self, tanggal, mood):
        self.data_mood[tanggal] = mood

    def lihat_mood(self, tanggal):
        return self.data_mood.get(tanggal, "Belum dicatat")

class Planner:
    def __init__(self):
        self.jadwal = []
        self.mood_tracker = MoodHarian()

    def tambah_kegiatan(self, nama_matkul, hari, jam_mulai, jam_selesai):
        kegiatan = KegiatanKuliah(nama_matkul, hari, jam_mulai, jam_selesai)
        self.jadwal.append(kegiatan)

    def tampilkan_jadwal(self):
        print("\nJadwal Kegiatan Kuliah:")
        for k in self.jadwal:
            print("-", k)

    def catat_mood_hari_ini(self, mood):
        hari_ini = datetime.date.today().isoformat()
        self.mood_tracker.catat_mood(hari_ini, mood)
        print(f"Mood hari ini ({hari_ini}) tercatat: {mood}")

    def lihat_mood_tanggal(self, tanggal):
        mood = self.mood_tracker.lihat_mood(tanggal)
        print(f"Mood pada {tanggal}: {mood}")

# Contoh penggunaan
planner = Planner()
planner.tambah_kegiatan("Bahasa Inggris", "Senin", "15:20", "17:50")
planner.tambah_kegiatan("Statistika", "Rabu", "09:00", "11:30")
planner.tampilkan_jadwal()

planner.catat_mood_hari_ini("semangat")
planner.lihat_mood_tanggal(datetime.date.today().isoformat())
```

```python
chatfrom datetime import datetime

class MoodTracker:
    moods = []

    def __init__(self, date, mood):
        self.date = date
        self.mood = mood
        MoodTracker.moods.append(self)

    @classmethod
    def mood_summary(cls):
        from collections import Counter
        mood_counts = Counter(m.mood for m in cls.moods)
        return dict(mood_counts)

    @staticmethod
    def is_valid_mood(mood):
        return mood.lower() in ['senang', 'sedih', 'biasa', 'marah', 'cemas']


class Schedule:
    all_schedules = []

    def __init__(self, day, time, activity):
        self.day = day.capitalize()
        self.time = time
        self.activity = activity
        Schedule.all_schedules.append(self)

    @staticmethod
    def format_activity(schedule):
        return f"{schedule.day} - {schedule.time}: {schedule.activity}"

    @classmethod
    def get_by_day(cls, day):
        return list(filter(lambda s: s.day.lower() == day.lower(), cls.all_schedules))


# Fungsi untuk menambahkan data menggunakan input
def tambah_jadwal():
    day = input("Hari (cth: Senin): ")
    time = input("Waktu (cth: 15:00): ")
    activity = input("Kegiatan: ")
    Schedule(day, time, activity)
    print("✅ Jadwal ditambahkan!\n")


def tambah_mood():
    date = input("Tanggal (YYYY-MM-DD): ")
    mood = input("Mood hari ini (senang/sedih/biasa/marah/cemas): ")
    if MoodTracker.is_valid_mood(mood):
        MoodTracker(date, mood)
        print("✅ Mood dicatat!\n")
    else:
        print("❌ Mood tidak valid.\n")


def tampilkan_jadwal():
    print("\n📅 Semua Jadwal:")
    hasil = map(Schedule.format_activity, Schedule.all_schedules)
    print("\n".join(hasil), "\n")


def tampilkan_jadwal_hari_tertentu():
    day = input("Masukkan hari: ")
    hasil = Schedule.get_by_day(day)
    if hasil:
        print("\n📅 Jadwal di hari", day.capitalize())
        for h in hasil:
            print(Schedule.format_activity(h))
    else:
        print("📭 Tidak ada jadwal pada hari itu.")
    print()


def tampilkan_mood_summary():
    print("\n📊 Rekap Mood:")
    for mood, count in MoodTracker.mood_summary().items():
        print(f"{mood.capitalize()}: {count}x")
    print()


# Menu interaktif sederhana
def main():
    while True:
        print("=== MENU UTAMA ===")
        print("1. Tambah Jadwal")
        print("2. Tambah Mood")
        print("3. Lihat Semua Jadwal")
        print("4. Lihat Jadwal per Hari")
        print("5. Lihat Rekap Mood")
        print("6. Keluar")
        pilih = input("Pilih opsi (1-6): ")

        match pilih:
            case "1": tambah_jadwal()
            case "2": tambah_mood()
            case "3": tampilkan_jadwal()
            case "4": tampilkan_jadwal_hari_tertentu()
            case "5": tampilkan_mood_summary()
            case "6": print("Bye 👋"); break
            case _: print("❌ Pilihan tidak valid.\n")


if __name__ == "__main__":
    main()
```
